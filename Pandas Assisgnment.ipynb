1. Finding the number of canceled flights for every airline per weekday
2. Finding the number and percentage of canceled and diverted flights for every airline per weekday. 
 
  For each origin and destination,
3. finding the total number of flights, the number and percentage of canceled flights, and the average and variance of the airtime.

```python
import pandas as pd

url = r'C:\Users\Admin\Desktop\flights-2.csv'

flight = pd.read_csv(url)

flight                         # To load the data
```

```python
flight.info()
```

```python
 #1. To find the number of canceled flights for every airline per weekday
    
    
flight.groupby("AIRLINE")['CANCELLED','WEEKDAY'].sum()
```

```python
 #2. To find the number and percentage of canceled and diverted flights for every airline per weekday.

flight.groupby("AIRLINE")['CANCELLED','DIVERTED', 'WEEKDAY'].sum()
```

```python
flight[["CANCELLED", "DIVERTED", "WEEKDAY"]].mean()
```

```python
#3. For each origin and destination,

finding the total number of flights, the number and percentage of canceled flights, and the average and variance of the airtime.
```

```python
flight.groupby("DEST_AIR")['CANCELLED', 'AIR_TIME', 'WEEKDAY'].sum()   # To find the number and average for each destination
```

```python
flight.groupby("ORG_AIR")['CANCELLED', 'AIR_TIME', 'WEEKDAY'].sum()     # To find the number and average for each origin
```

```python
link = r'C:\Users\Admin\Desktop\covid_subset-1.txt'
covid = pd.read_csv(link)
covid                                     # To load the data
```

```python
# To select the some columns and save in a new dataframe

covid19 = covid[["continent", "location", "total_cases", "total_deaths", "gdp_per_capita", "tests_per_case",
                 "life_expectancy", "female_smokers","male_smokers", "diabetes_prevalence"]]

covid19
```

```python
 # To Find the missing values in covid19

covid19.isna().sum()      # There are 205 missing values in tests_per_case, female_smokers and male_smokers respectively
```

```python
# To devise an appropriate approach with reasonable explanation on how to handle the missing values.

covid19[["tests_per_case", "female_smokers", "male_smokers"]]
```

##### There are 205 rows in the data (as shown from the dimension of the data). The missing values above also indicate there are empty values under tests_per_case, female_smokers and male_smokers respectively. This means no values were supplied. My strategy will be to drop these columns. 

```python
  
covid19.dropna(axis = 1)
```

```python
covid19.isna().sum()
```

```python
# To find the Continent with the highest total cases and total deaths of coronavirus

covid19.groupby('continent')['total_cases','total_deaths'].max()  
```

```python
 # To find the Country/location with the highest number of deaths

covid19.groupby('location')['total_deaths'].max()
```

```python
# To find the Continent with the highest number of deaths and total cases of Covid-19

covid19.groupby('continent')['total_deaths','total_cases'].max()
```
